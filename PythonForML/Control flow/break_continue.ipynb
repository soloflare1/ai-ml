## continue


```python
sum = 0

for i in range(1,101):
    
    print (i, " is processing")
    
    if i % 2 == 0: # even
       continue    # skip
   
    sum += i
    print(i," is added to the sum")
    
print(sum)
```

## Break

ml -> model train -> accuracy -> x% increases, and if it reached to 100% then stop(break)

by for loop


```python
accuracy = 30

for i in range(20):
    accuracy += 1
    print(accuracy)
    
    if accuracy == 100:
        break
    
```

by while loop

```python
accuracy = 30

while accuracy < 100:
    accuracy += 1
    print(accuracy)
    
    if accuracy == 100:
        break
```
