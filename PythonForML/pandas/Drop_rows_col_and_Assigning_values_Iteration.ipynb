## value assign and change


```python
# df.loc[idx, 'col name']

#  1 no idx er student er marks 90 korbo and 
df.loc[1, 'Python Marks'] = 90

df.loc[1,'CompletionStatus'] = 'Completed'

df.head(5)
```

```python
# .loc[1:3, 'Python Marks'] → selects rows 1 through 3 (inclusive!) 
# and the column 'Python Marks'.
# += 2 → adds 2 marks to those rows in that column.

df.loc[1:3, 'Python Marks'] += 2
df.head()
```

```python
for i, series in df.iterrows():
    print(f"{i} : {series}")
```
